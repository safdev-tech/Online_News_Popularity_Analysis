```python
#Data loading 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
Online_Popularity = pd.read_csv("OnlineNewsPopularity.csv")
Online_Popularity.head()
```

```python
Online_Popularity.info()
```

```python
Online_Popularity['shares'].mean()
```

```python
Online_Popularity.describe()
```

```python
Online_Popularity = Online_Popularity.drop(['url'], axis = 1)
Online_Popularity.head()
```

```python
y = Online_Popularity['shares']
y
```

```python
x = Online_Popularity.drop(['shares'], axis = 1)
x
```

```python
Online_Popularity.corr()
```

```python
sns.heatmap(Online_Popularity.corr())
```

```python
# Train test splitting
from sklearn.model_selection import train_test_split
```

```python
trainx, testx, trainy, testy = train_test_split(x, y, test_size=0.3)
```

```python
trainx.shape
```

```python
testx.shape
```

```python
trainy.shape
```

```python
testy.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(trainx, trainy)
```

```python
Predicted_Y = lm.predict(testx)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(testy, Predicted_Y)
```

```python
plt.scatter(testy, Predicted_Y)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')

plt.ylim(0,15000)
plt.xlim(0,65000)
```
